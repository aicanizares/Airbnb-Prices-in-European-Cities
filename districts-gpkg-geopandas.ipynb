```python
# Convert Airbnb dataset to a GeoDataFrame
df_airbnb["geometry"] = df_airbnb.apply(lambda row: Point(row["lng"], row["lat"]), axis=1)
gdf_airbnb = gpd.GeoDataFrame(df_airbnb, geometry="geometry", crs="EPSG:4326")
```

```python
# Define cities and their corresponding countries
city_country_map = {
    "Amsterdam": "Netherlands",
    "Athens": "Greece",
    "Barcelona": "Spain",
    "Berlin": "Germany",
    "Budapest": "Hungary",
    "Lisbon": "Portugal",
    "London": "United Kingdom",
    "Paris": "France",
    "Rome": "Italy",
    "Vienna": "Austria"
}

# Filter districts where NAME_4 matches a city AND NAME_0 matches the correct country
gdf_filtered = gdf[gdf.apply(lambda row: row["NAME_4"] in city_country_map and row["NAME_0"] == city_country_map[row["NAME_4"]], axis=1)]
```

```python
# Ensure both datasets have the same CRS
gdf_filtered = gdf_filtered.to_crs(gdf_airbnb.crs)

# Perform the spatial join (Airbnb listings WITH districts)
gdf_joined = gpd.sjoin(gdf_airbnb, gdf_filtered, how="left", predicate="within")
```

```python
# Keep relevant columns (Replace "district_column" with the actual column name in your dataset)
df_airbnb["district_name"] = gdf_joined["NAME_4"]
df_airbnb["district_lat"] = gdf_joined.geometry.centroid.y  # District's latitude
df_airbnb["district_lng"] = gdf_joined.geometry.centroid.x  # District's longitude
```
